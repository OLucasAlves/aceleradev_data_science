```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Lucas\codenation\coestatistica-1\desafio1.csv') 
```

```python
df.dtypes
```

```python
df_submission = pd.DataFrame(df.groupby('estado_residencia')['pontuacao_credito'].mean())
df_submission
```

```python
df_submission.rename(columns = {'pontuacao_credito' : 'media'}, inplace = True)
df_submission
```

```python
df_submission['mediana'] = df.groupby('estado_residencia')['pontuacao_credito'].median()
df_submission['desvio_padrao'] = df.groupby('estado_residencia')['pontuacao_credito'].std()
df_submission['moda'] = df.groupby('estado_residencia')['pontuacao_credito'].agg(pd.Series.mode)
df_submission
```

```python
df_submission = df_submission.transpose()
df_submission
```

```python
df_submission.to_json('submission.json')
```
